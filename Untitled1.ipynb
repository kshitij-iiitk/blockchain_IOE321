```python
import hashlib

# -------------------------------
# SHA-256 hash function
# -------------------------------
def sha256(data):
    return hashlib.sha256(data.encode('utf-8')).hexdigest()


# -------------------------------
# Merkle Tree Construction
# -------------------------------
def build_merkle_tree(transactions):

    print("\nðŸ”¹ Step 1: Hashing Transactions (Leaf Nodes)")
    level = 0
    hashes = [sha256(tx) for tx in transactions]

    for i, h in enumerate(hashes):
        print(f"  Tx{i+1} Hash: {h}")

    # Build tree levels
    while len(hashes) > 1:
        print(f"\nðŸ”¹ Step 2: Building Level {level + 1}")
        temp = []

        # If odd number of nodes, duplicate last hash
        if len(hashes) % 2 != 0:
            print("  Odd number of hashes â†’ duplicating last hash")
            hashes.append(hashes[-1])

        # Pairwise hashing
        for i in range(0, len(hashes), 2):
            combined = hashes[i] + hashes[i+1]
            parent_hash = sha256(combined)
            temp.append(parent_hash)

            print(f"  Hash({hashes[i][:8]}..., {hashes[i+1][:8]}...)")
            print(f"   â†’ {parent_hash}")

        hashes = temp
        level += 1

    print("\nðŸ”¹ Final Step: Merkle Root")
    print(f"  Merkle Root: {hashes[0]}")

    return hashes[0]


# -------------------------------
# Main Execution
# -------------------------------
if __name__ == "__main__":

    print(" Merkle Tree Construction Demo (Blockchain)")
    print("-------------------------------------------")

    transactions = [
        "Alice pays Bob 10 BTC",
        "Bob pays Charlie 5 BTC",
        "Charlie pays Dave 2 BTC",
        "Dave pays Eve 1 BTC"
    ]

    merkle_root = build_merkle_tree(transactions)
```
